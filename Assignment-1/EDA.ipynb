# Exploratory Data Analysis

This notebook will examine the distributions of each of the desired features specified in encoded_dirty.csv

```python
import pandas as pd
import numpy as np
import plotly.express as px
import matplotlib.pyplot as plt

df = pd.read_csv('encoded_dirty.csv')               # importing the dataframe
df.drop(inplace=True, labels='Unnamed: 0', axis=1)  # dropping an unnecessary index column
```

```python
df.columns
```

```python
import seaborn as sns
```

```python
sns.distplot(df['resale_year'])
plt.show()
```

```python
fig, axes = plt.subplots(3,3, figsize=(30,15))
sns.distplot(df['resale_year'],ax=axes[0,0], kde_kws=dict(bw=0.5))
sns.distplot(df['resale_month'],ax=axes[0,1], kde_kws=dict(bw=0.5))
sns.distplot(df['flat_type'],ax=axes[0,2], kde_kws=dict(bw=0.5))
sns.distplot(df['floor_area_sqm'],ax=axes[1,0], kde_kws=dict(bw=0.5))
sns.distplot(df['lease_commence_date'],ax=axes[1,1], kde_kws=dict(bw=0.5))
sns.distplot(df['lat'],ax=axes[1,2], kde_kws=dict(bw=0.5))
sns.distplot(df['long'],ax=axes[2,0], kde_kws=dict(bw=0.5))
sns.distplot(df['remaining_lease'],ax=axes[2,1], kde_kws=dict(bw=0.5))
sns.distplot(df['storey_range'],ax=axes[2,2], kde_kws=dict(bw=0.5))
```

```python
sns.distplot(df['long'], kde_kws=dict(bw=1.0))
```

```python
sns.distplot(df['floor_area_sqm'], kde_kws=dict(bw=0.7))
```

```python
sns.distplot(df['resale_year'], kde_kws=dict(bw=0.8))
```

```python
fig, axes = plt.subplots(3,3, figsize=(30,15))
sns.boxplot(df['resale_year'],ax=axes[0,0])
sns.boxplot(df['resale_month'],ax=axes[0,1])
sns.boxplot(df['flat_type'],ax=axes[0,2])
sns.boxplot(df['floor_area_sqm'],ax=axes[1,0])
sns.boxplot(df['lease_commence_date'],ax=axes[1,1])
sns.boxplot(df['lat'],ax=axes[1,2])
sns.boxplot(df['long'],ax=axes[2,0])
sns.boxplot(df['remaining_lease'],ax=axes[2,1])
sns.boxplot(df['storey_range'],ax=axes[2,2])
```

From the boxplots above, we can see that 3 features have a significant number of outliers:
1. floor_area_sqm [1,0]
2. remaining_lease [2,1]
3. storey_range [2,2]
